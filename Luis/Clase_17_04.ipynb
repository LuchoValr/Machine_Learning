##Proyecto Ajuste Polinomial

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

#Establecemos las variables
x = np.array([0,1,2,3,4,5])
y = np.array([0,0.8,0.9,0.1,-0.8,-1.0])
```

```python
#polyfit se utiliza  para ajustar un conjunto de datos
pol = np.polyfit(x,y,1)
pol
#Ecuaci√≥n y = -0.30x+0.75
```

```python
#Calcular el polinomio, utilizamos np.polyval
y_cor = np.polyval(pol,x)
y_cor
```

```python
#Graficamos
plt.plot(x, y, "o",y_cor, "o--")
plt.show()
#datos en azul reales, pronoticados hacia atras naranja, probabilidad de datos errados hacia adelante es errados dado que estan muy alejados de los datos reales
```

```python
#calculamos R2
sse = sum((y - y_cor)**2)
tse = (len(y)-1) * np.var(y)
r2 = 1-sse/tse
print(r2)
```

```python
#Regresi√≥n cuadr√°tica
pol2 = np.polyfit(x, y, deg=2 ) #grado de polinomio
pol2
# Ecuaci√≥n polinomica  y = -0.16x^2+0.50x+0.22
```

```python
#validamos
y_cor2 = np.polyval(pol2, x)
y_cor2
```

```python
#Graficamos
plt.plot(x, y, "o--", y_cor2, "*--" )
plt.show()
```

```python
#R2 nuevo
sse = sum((y - y_cor2)**2)
tse = (len(y)-1) * np.var(y)
r22 = 1-sse/tse
print(r22)
```

```python
#Regresi√≥n 
ypol3 = np.polyfit(x, y, deg= 3)
ypol3
#Ecuaci√≥n 0.087x^3-0.81^2+1.69x-0.03
```

```python
y_cor3 = np.polyval(ypol3, x)
y_cor3
```

```python
sse = sum((y - y_cor3)**3)
tse = (len(y)-1) * np.var(y)
r23 = 1-sse/tse
print(r23)
```
